# Challenge 3 - Handle Missing Values

The next step would be handling missing values. **We start by examining the number of missing values in each column, which you will do in the next cell.**

```python
missing_counts = X_droped.isnull().sum()
print(f"Missing Values Count: {missing_counts[missing_counts>0]}")
```

```python
def clean_missing_values(df, threshold=0.5):
    if df.empty: return df
    x_len = len(df)
    missing_counts = df.isnull().sum()
    columns_more_than_threshold = missing_counts[missing_counts > x_len * threshold]
    if columns_more_than_threshold.empty:
        print(f"No columns have more than {threshold*100}% missing values. No columns will be dropped based on this criterion.")
    else:
        columns_to_drop = columns_more_than_threshold.index
        df = df.drop(columns=columns_to_drop)
        print(f"Columns dropped due to missing values: {list(columns_to_drop)}")
    return df
```

```python
# decision to drop columns with more than 44%, since we have one column with 812 missing values, loosing 45% of that data will affect more than drop a column with less than 0.05 correlation due Xgboost feature importance.
# decision base on knowledge about the malicious servers hide this content so this data could not reflect reality.
X_droped = clean_missing_values(X_droped, threshold=0.44)
```

```python
# Decision: We will fill the missing values in 'WHOIS_UPDATED_DATE' with the values from 'WHOIS_REGDATE' because they are likely to be related
# (the update date can't be before the registration date). This way we can preserve more data for our model.
X_droped['WHOIS_UPDATED_DATE'] = X_droped['WHOIS_UPDATED_DATE'].fillna(X_droped['WHOIS_REGDATE'])
```

```python
# 1. Fill categorical columns where missing data has meaning (Anonymity)
categorical_to_fill = ['SERVER', 'WHOIS_COUNTRY', 'WHOIS_STATEPRO']
for col in categorical_to_fill:
    X_droped[col] = X_droped[col].fillna('unknown')

# 2. Drop the rows with missing dates and DNS_QUERY_TIMES
# We decided to drop these because handling date strings is complex for an initial ML exercise
initial_rows = len(X_droped)
X_droped = X_droped.dropna()
final_rows = len(X_droped)

print(f"Rows removed (Dates & DNS nulls): {initial_rows - final_rows}")
print(f"Final dataset shape: {X_droped.shape}")
print(f"Remaining missing values: {X_droped.isnull().sum().sum()}")
```

Not yet. If you reflect on the previous labs how we handle categorical columns, you probably remember we ended up dropping a lot of those columns because there are too many unique values. Too many unique values in a column is not desirable in machine learning because it makes prediction inaccurate. But there are workarounds under certain conditions. One of the fixable conditions is:

#### If a limited number of values account for the majority of data, we can retain these top values and re-label all other rare values.

The `WHOIS_COUNTRY` column happens to be this case. You can verify it by print a bar chart of the `value_counts` in the next cell to verify:

```python
def print_bar_plot(df, column_name):
    
    sns.set_theme(style="dark")
    data = df[column_name].value_counts()
    colors = sns.color_palette("viridis", n_colors=len(data))

    
    fig, ax = plt.subplots(figsize=(12, 7))
    fig.set_facecolor('#212529')
    ax.set_facecolor('#212529')

    ax.bar(data.index, data.values, color=colors, alpha=1.0)

    ax.tick_params(colors='grey')
    ax.xaxis.label.set_color('grey')
    ax.yaxis.label.set_color('grey')
    ax.title.set_color('grey')

    for spine in ax.spines.values():
        spine.set_edgecolor('grey')

    plt.title(f'{column_name} by Frequency', fontsize=15, pad=20)
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    
    plt.tight_layout()
    plt.show()
```

```python
print_bar_plot(X_droped, 'WHOIS_COUNTRY')
```

```python
def regularize_dates(df, reg_col='WHOIS_REGDATE', upd_col='WHOIS_UPDATED_DATE'):
    """
    Complete date regularization for cybersecurity datasets.
    Standardizes formats (DD/MM/YYYY), handles missing values, 
    and generates temporal features.
    """
    df = df.copy()

    # 1. Standardizing to Datetime objects
    # dayfirst=True ensures 31/12 is not read as Month 31
    # errors='coerce' turns unparseable strings into NaT (Not a Time)
    df[reg_col] = pd.to_datetime(df[reg_col], dayfirst=True, errors='coerce')
    df[upd_col] = pd.to_datetime(df[upd_col], dayfirst=True, errors='coerce')

    # 2. Regularization Logic: Repairing missing data
    # If Updated Date is missing but Registration exists, assume Update = Registration
    df[upd_col] = df[upd_col].fillna(df[reg_col])
    
    # If Registration is missing but Updated exists, assume Registration = Update
    df[reg_col] = df[reg_col].fillna(df[upd_col])

    # 3. Handling remaining Nulls (cases where both are missing)
    # Instead of dropping, we fill with the median to preserve the row 
    # but we mark them so the model knows the data was missing
    df['DATE_DATA_MISSING'] = df[reg_col].isnull().astype(int)
    
    median_reg = df[reg_col].median()
    median_upd = df[upd_col].median()
    
    df[reg_col] = df[reg_col].fillna(median_reg)
    df[upd_col] = df[upd_col].fillna(median_upd)

    # 4. Feature Extraction (The "meat" for the models)
    df['REG_YEAR'] = df[reg_col].dt.year
    df['REG_MONTH'] = df[reg_col].dt.month
    df['UPD_YEAR'] = df[upd_col].dt.year
    df['UPD_MONTH'] = df[upd_col].dt.month
    
    # 5. Calculated Metric: Domain Longevity/Age
    # High longevity usually correlates with benign websites
    df['DOMAIN_AGE_DAYS'] = (df[upd_col] - df[reg_col]).dt.days
    
    # Ensuring no negative ages due to inconsistent data entry
    df['DOMAIN_AGE_DAYS'] = df['DOMAIN_AGE_DAYS'].apply(lambda x: x if x >= 0 else 0)

    # 6. Cleanup: Remove the original datetime objects
    df = df.drop(columns=[reg_col, upd_col])

    return df
```

```python
# --- 1. PREPROCESSING FUNCTIONS ---

def clean_missing_values(df, threshold=0.6):
    """Drops columns with more missing values than the threshold."""
    if df.empty: return df
    x_len = len(df)
    missing_counts = df.isnull().sum()
    columns_to_drop = missing_counts[missing_counts > x_len * threshold].index
    if not columns_to_drop.empty:
        df = df.drop(columns=columns_to_drop)
        print(f"Dropped columns: {list(columns_to_drop)}")
    return df

def consolidate_server(server_name):
    """Standardizes server names into categories."""
    server_name = str(server_name).lower()
    if 'microsoft' in server_name: return 'microsoft'
    elif 'apache' in server_name: return 'apache'
    elif 'nginx' in server_name: return 'nginx'
    else: return 'other'

def country_codes(df, column):
    """Maps country codes and fills unknowns."""
    good_country = {
        'None':'None', 'unknown':'Unknown', 'US':'US', 'SC':'SC', 'GB':'UK', 
        'UK':'UK', 'RU':'RU', 'AU':'AU', 'CA':'CA', 'PA':'PA', 'se':'SE', 
        'IN':'IN', 'LU':'LU', 'TH':'TH', "[u'GB'; u'UK']":'UK', 'FR':'FR',
        'NL':'NL', 'UG':'UG', 'JP':'JP', 'CN':'CN', 'SE':'SE', 'SI':'SI', 
        'IL':'IL', 'ru':'RU', 'KY':'KY', 'AT':'AT', 'CZ':'CZ', 'PH':'PH', 
        'BE':'BE', 'NO':'NO', 'TR':'TR', 'LV':'LV', 'DE':'DE', 'ES':'ES', 
        'BR':'BR', 'us':'US', 'KR':'KR', 'HK':'HK', 'UA':'UA', 'CH':'CH', 
        'United Kingdom':'UK', 'BS':'BS', 'PK':'PK', 'IT':'IT', 
        'Cyprus':'CY', 'BY':'BY', 'AE':'AE', 'IE':'IE', 'UY':'UY', 'KG':'KG'
    }
    df[column] = df[column].apply(lambda x : good_country.get(x, 'Unknown'))
    return df

def regularize_dates(df, reg_col='WHOIS_REGDATE', upd_col='WHOIS_UPDATED_DATE'):
    """Regularizes dates using European format and calculates domain metrics."""
    # Explicitly parsing dayfirst=True for European format
    df[reg_col] = pd.to_datetime(df[reg_col], dayfirst=True, errors='coerce')
    df[upd_col] = pd.to_datetime(df[upd_col], dayfirst=True, errors='coerce')

    # Regularization: Cross-filling missing dates
    df[upd_col] = df[upd_col].fillna(df[reg_col])
    df[reg_col] = df[reg_col].fillna(df[upd_col])

    # Handling remaining NaNs with median to preserve rows
    median_reg = df[reg_col].median()
    median_upd = df[upd_col].median()
    df[reg_col] = df[reg_col].fillna(median_reg)
    df[upd_col] = df[upd_col].fillna(median_upd)

    # Feature Extraction
    df['REG_YEAR'] = df[reg_col].dt.year
    df['REG_MONTH'] = df[reg_col].dt.month
    df['DOMAIN_AGE_DAYS'] = (df[upd_col] - df[reg_col]).dt.days
    df['DOMAIN_AGE_DAYS'] = df['DOMAIN_AGE_DAYS'].apply(lambda x: x if x >= 0 else 0)

    # Drop original date columns
    df = df.drop(columns=[reg_col, upd_col])
    return df

# --- 2. MAIN PIPELINE ---

# Load
df = pd.read_csv('../website.csv')

# Step 1: Handle Missing Values (Keep CONTENT_LENGTH by setting threshold at 0.6)
df = clean_missing_values(df, threshold=0.6)

# Step 2: Regularize Dates (European format + Feature Engineering)
df = regularize_dates(df)

# Step 3: Categorical Treatment
df['SERVER'] = df['SERVER'].apply(consolidate_server)
df = country_codes(df, 'WHOIS_COUNTRY')
df['CHARSET'] = df['CHARSET'].fillna('unknown')

# Step 4: Numerical Imputation (for columns like CONTENT_LENGTH)
df['CONTENT_LENGTH'] = df['CONTENT_LENGTH'].fillna(df['CONTENT_LENGTH'].median())
df['DNS_QUERY_TIMES'] = df['DNS_QUERY_TIMES'].fillna(df['DNS_QUERY_TIMES'].median())

# Step 5: Encoding
#df = pd.get_dummies(df, columns=['CHARSET', 'SERVER', 'WHOIS_COUNTRY'], drop_first=True)
df = pd.get_dummies(df, columns=['CHARSET', 'SERVER'], drop_first=True)
print(f"Final dataset size: {df.shape[0]} rows")
print(f"Total features after processing: {df.shape[1]}")
```
